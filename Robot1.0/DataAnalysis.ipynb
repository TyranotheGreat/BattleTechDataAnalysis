# Weapons fired overall

```python
%%sql
select weapon, count(*) thecount
from blahbig
group by weapon
order by thecount desc
```

# Weapons fired by mech and weapon

```python
%%sql
select attacking_mechdef_id, weapon, count(*) thecount
from blahbig
group by attacking_mechdef_id, weapon
order by thecount desc
```

# Mechs with the most wins (The last one to fire)

```python
%%sql
select attacking_mechdef_id, defending_mechdef_id, count(*) wins
from 
(
select attacking_mechdef_id, defending_mechdef_id, row_number() over (partition by battle_id order by round desc, turn desc) therownum
from blahbig
) a
where a.therownum = 1
group by attacking_mechdef_id, defending_mechdef_id
order by wins desc
```

# Mechs with the most losses (The last one fired upon)

```python
%%sql
select defending_mechdef_id, count(*) losses
from 
(
select attacking_mechdef_id, defending_mechdef_id, row_number() over (partition by battle_id order by round desc, turn desc) therownum
from blahbig
) a
where a.therownum = 1
group by defending_mechdef_id
order by losses desc
```

# Average, max, and min number of rounds to end a battle

```python
%%sql
select min(round) quickest, max(round) longest, avg(round) average
from 
(
select battle_id, round, turn, row_number() over (partition by battle_id order by round desc, turn desc) therownum
from blahbig
) a
where a.therownum = 1
```

# Biggest, smallest, and average damage done per battle per mech

```python
%%sql
select attacking_mechdef_id, min(total_damage) min_damage, max(total_damage) max_damage, cast(avg(total_damage) as int) average_damage
from 
(
select battle_id, attacking_mechdef_id, sum(target_damage_taken) total_damage
from blahbig
group by battle_id, attacking_mechdef_id
) a
group by attacking_mechdef_id
order by attacking_mechdef_id
```
